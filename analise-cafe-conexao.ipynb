```python
 # Passo 1: Importar as bibliotecas necess√°rias
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
data = {
    'Categoria': ['Caf√©s Especiais', 'Salgados', 'Doces e Sobremesas'],
    'Receita Total (R$)': [120000, 95000, 65000],
    'Itens Vendidos': [8000, 11875, 5200],
    'Custo dos Ingredientes (R$)': [30000, 47500, 16250],
    'Clientes √önicos': [2500, 4750, 3250]
}

df = pd.DataFrame(data)
```

```python
# --- AN√ÅLISE E C√ÅLCULOS ---

# Passo 3: C√°lculos de Rentabilidade e Efici√™ncia
# Calcular Lucro Bruto (Receita - Custo)
df['Lucro Bruto (R$)'] = df['Receita Total (R$)'] - df['Custo dos Ingredientes (R$)']

# Calcular Margem de Lucro Bruta em porcentagem
df['Margem de Lucro (%)'] = (df['Lucro Bruto (R$)'] / df['Receita Total (R$)']) * 100

# Calcular Ticket M√©dio por Item (pre√ßo m√©dio por item vendido)
df['Ticket M√©dio (R$)'] = df['Receita Total (R$)'] / df['Itens Vendidos']
```

```python
# --- EXIBI√á√ÉO DOS RESULTADOS NO CONSOLE ---

print("="*50)
print("An√°lise de Dados da Cafeteria 'Caf√© & Conex√£o'")
print("="*50)
print("\n")

# Exibir a tabela completa com os novos c√°lculos
print("Tabela de An√°lise Completa:")
# Usamos .round(2) para formatar os n√∫meros com 2 casas decimais
print(df.round(2).to_string())
print("\n")
```

```python
# Exibir os destaques da an√°lise
print("--- Destaques da An√°lise ---")
receita_max = df.loc[df['Receita Total (R$)'].idxmax()]
lucro_max = df.loc[df['Lucro Bruto (R$)'].idxmax()]
print(f"üèÜ Categoria com MAIOR RECEITA: {receita_max['Categoria']} (R$ {receita_max['Receita Total (R$)']:,})")
print(f"üí∞ Categoria com MAIOR LUCRO BRUTO: {lucro_max['Categoria']} (R$ {lucro_max['Lucro Bruto (R$)']:,})")
print("\n")
```

```python
# Configura√ß√µes de estilo para os gr√°ficos
plt.style.use('seaborn-v0_8-whitegrid')

# Gr√°fico 1: Comparativo de Lucro Bruto
# Ordenar os dados para um gr√°fico mais claro
df_sorted_lucro = df.sort_values('Lucro Bruto (R$)', ascending=False)

plt.figure(figsize=(10, 6)) # Define o tamanho da imagem
bars_lucro = plt.bar(df_sorted_lucro['Categoria'], df_sorted_lucro['Lucro Bruto (R$)'], color=['#004c6d', '#6996b3', '#b3deff'])
plt.title('Lucro Bruto por Categoria', fontsize=16, fontweight='bold')
plt.ylabel('Lucro Bruto (R$)', fontsize=12)
plt.xticks(rotation=0, fontsize=11)

# Adicionar os valores no topo de cada barra
for bar in bars_lucro:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'R$ {int(yval):,}', va='bottom', ha='center', fontsize=11)

plt.tight_layout() # Ajusta o layout para evitar cortes
plt.savefig('grafico_lucro_bruto.png') # Salva o gr√°fico em um arquivo
```

```python
# Gr√°fico 2: Comparativo de Margem de Lucro
df_sorted_margem = df.sort_values('Margem de Lucro (%)', ascending=False)

plt.figure(figsize=(10, 6))
bars_margem = plt.bar(df_sorted_margem['Categoria'], df_sorted_margem['Margem de Lucro (%)'], color=['#006837', '#78c679', '#ccece6'])
plt.title('Margem de Lucro (%) por Categoria', fontsize=16, fontweight='bold')
plt.ylabel('Margem de Lucro (%)', fontsize=12)
plt.xticks(rotation=0, fontsize=11)

# Adicionar os valores no topo de cada barra
for bar in bars_margem:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.0f}%', va='bottom', ha='center', fontsize=11)

plt.tight_layout()
plt.savefig('grafico_margem_de_lucro.png')
```

```python
# Exibe os gr√°ficos na tela ao final da execu√ß√£o do script
plt.show()

print("An√°lise conclu√≠da. Gr√°ficos 'grafico_lucro_bruto.png' e 'grafico_margem_de_lucro.png' foram salvos.")
```
